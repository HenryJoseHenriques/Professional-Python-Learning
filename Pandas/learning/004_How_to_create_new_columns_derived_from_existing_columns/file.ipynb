## Como crio uma nova coluna?

Para criar uma nova coluna, vc usa [] com o nome da nova coluna que deseja adicionar.

```python
population["Population_2025_Expectative"] = population["Population_2024"] * 1.1

population.head()
```

Gostaria de verificar a diferença da população de 2023 para a de 2024 e armazenar os valores em uma coluna

```python
population["Growth Population 2023-2024"] = population["Population_2024"] - population["Population_2023"]
population.head()
```

As demais operações matemáticas e operações lógicas funcionam elemento-a-elemento no pandas

rename() => é possível renomear as colunas utilizando o método rename, porém é necessário passa um dicionário como parâmetro

```python
population_rename = population.rename(
    columns={
        "Population_2023": "2023_Population",
        "Population_2024": "2024_Population",
        "Population_2025_Expectative": "2025_Expectative_Population"
    }
)

population_rename.head()
```

A correspondência não deve se restringir apenas a nomes fixos, mas também pode ser uma função de mapeamento. Por exemplo, converter os nomes das colunas para letras minúsculas pode ser feito usando uma função:

```python
population_rename = population_rename.rename(columns=str.upper)
#population_rename = population_rename.rename(columns=str.lower)

population_rename.head()
```

## Lembre-se 

* Crie uma nova coluna atribuindo a saída ao DataFrame com um novo nome de coluna entre os colchetes [].
* As operações são realizadas elemento por elemento, não é necessário fazer um loop sobre as linhas.
* Use rename com um dicionário ou função para renomear rótulos de linhas ou nomes de colunas.
