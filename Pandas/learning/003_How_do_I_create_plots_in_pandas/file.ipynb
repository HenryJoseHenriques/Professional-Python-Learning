```python
import pandas as pd
import matplotlib.pyplot as plt

studant_performace = pd.read_csv("../../dataFrames/student_performance_data.csv", index_col=0, parse_dates=True)
#studant_performace = pd.read_csv("../../dataFrames/student_performance_data.csv", index_col = 0)

studant_performace.head()
```

plot() -> imprimi graficamente uma imagem. Pode ser usados para DataFrames e Series.

```python
studant_performace["AttendanceRate"].plot()

plt.show()
```

## gr√°fico de dispers√£o

√â poss√≠vel dispor em um gr√°fico de dispers√£o, utilizando o m√©todo scatter(x=,y=,alpha=).

```python
studant_performace.plot.scatter(x="StudyHoursPerWeek",y="GPA", alpha = 0.5)
```

### Como interpretar gr√°ficos de dispers√£o

Eixo X e Eixo Y:

Cada ponto no gr√°fico representa um par de valores. O valor da vari√°vel no eixo X √© representado pela posi√ß√£o horizontal do ponto, e o valor da vari√°vel no eixo Y √© representado pela posi√ß√£o vertical do ponto.
Correla√ß√£o:

Correla√ß√£o positiva: Se os pontos tendem a subir da esquerda para a direita, significa que as duas vari√°veis t√™m uma correla√ß√£o positiva. Isso indica que, √† medida que uma vari√°vel aumenta, a outra tende a aumentar tamb√©m.
Correla√ß√£o negativa: Se os pontos tendem a descer da esquerda para a direita, h√° uma correla√ß√£o negativa. Isso indica que, √† medida que uma vari√°vel aumenta, a outra tende a diminuir.
Sem correla√ß√£o: Se os pontos est√£o espalhados de forma aleat√≥ria, sem formar um padr√£o claro, isso sugere que n√£o h√° correla√ß√£o aparente entre as vari√°veis.
For√ßa da Correla√ß√£o:

Se os pontos est√£o muito pr√≥ximos de uma linha reta, a correla√ß√£o √© forte.
Se os pontos est√£o mais dispersos, a correla√ß√£o √© mais fraca.

```python
desempenho = studant_performace.plot.area(figsize=(12,4), subplots = True )
#desempenho = studant_performace.plot.area(figsize=(12,4), subplots = False )
#desempenho = studant_performace.plot.area(figsize=(24,8), subplots = True )

plt.show()
```

Quero personalizar, estender ou salvar o gr√°fico resultante. 

```python
fig, axs = plt.subplots(figsize = (12,4))        #Cria eixos e figuras matploitlib vazias 
studant_performace.plot.area(ax=axs)            #Usa o pandas para colocar √°rea do gr√°fico nas figuras e eixos j√° preparados
axs.set_ylabel("Studante Performance")          #Ajusrta o nome do eixo vertical
fig.savefig("Grafico_de_performace_de_estudantes.png") #Salva o gr√°fico em uma imagem png
plt.show() 
```
