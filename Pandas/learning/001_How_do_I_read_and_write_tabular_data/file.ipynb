## Como eu leio e escrevo em uma tabela de dados?

![](https://pandas.pydata.org/docs/_images/02_io_readwrite.svg)

```python
import pandas as pd

sell = pd.read_excel("../../dataFrames/sales_base.xlsx") # ../.. garante que vc suba dois níveis
#naturalmente, existe também leitura de arquivos json, csv, html, SQl, dentre outros...

sell
```

head() -> é possível ler N primeras linhas

```python
sell.head(8)
```

tail() -> ler as últimas N linhas

```python
sell.tail(10)
```

dtypes -> atributo que verifica como o pandas interpretou cada dado da coluna

```python
sell.dtypes
```

## Armazenar dados

Assim como read_* é usado para ler dados, o to_* é usado para salvar dados, segue o exemplo:

```python
sell.to_csv("sell.csv", index = False) #a tabela sell.csv é armazenada dentro da pasta
```

info() -> fornece informações técnicas do dataFrame

```python
sell.info()
```

## Lembre-se

* A importação de dados em várias fontes é suportada pela funções 'read_*'.
* A exportação de dados em pandas por diferentes métodos 'to_*'.
* Os métodos 'head', 'tail', 'info' e o atributo dtypes são convenientes para uma primeira verificação.
