## Como selecionar um subconjunto de um DataFrame?

```python
import pandas as pd

death = pd.read_csv("../../dataFrames/Death_rates_for_suicide__by_sex__race__Hispanic_origin__and_age__United_States.csv")

death.head()
```

### Como seleciono uma coluna específica de uma dataFrame?
![](https://pandas.pydata.org/docs/_images/03_subset_columns.svg)

Seguindo o modelo abaixo, você consegue extrair uma coluna de interesse de um dataFrame

```python
year = death["YEAR"]

year
#year.head()
```

É possível também extrair mais que uma coluna

```python
year_num = death[["YEAR","YEAR_NUM"]] # não esqueça das [[]]
year_num
```

### Nota:

As colchetes internos definem uma lista Python com nomes de colunas, enquanto os colchetes externos são usados para selecionar os dados de um DataFrame do pandas, como visto no exemplo anterior.

### shape -> Atributo que retorna a quantidade de linhas e colunas

```python
death[["YEAR","YEAR_NUM"]].shape
```

## Como eu filtro linhas específicas de um DataFrame?

![](https://pandas.pydata.org/docs/_images/03_subset_rows.svg)


As expressões condicionais(< , > , <= , >= , !=) funcionam com pandas

```python
maior_2000 = death[death["YEAR"] > 2000]

maior_2000
```

```python
maior_2000.shape
```

Para selecionar linhas baseadas em uma condição, use a condição dentro da seleção dos colchetes




É possível verificar uma expressão booleana de uma coluna, devolvendo o condição com o número de linhas da coluna original que está sendo avaliada

```python
death["YEAR"] > 2000
```

```python
studant = pd.read_csv("../../dataFrames/student_performance_data.csv")

studant.head()
```

isin() -> retorna verdadeiro para cada linha que contém o valor presente na lista. é útil para filtros

```python
Major = studant[studant["Major"].isin(["Science","Arts"])]

Major
```

```python
GPA_2_3 = studant[(studant["GPA"] > 2) & (studant["GPA"] <= 3)]

GPA_2_3
```

### Nota: 
Ao combinar múltiplas declarações condicionais, cada condição deve estar rodeada por parênteses (). Além disso, você não pode usar or/and, mas precisa usar o operador | para ou e o operador & para e.


notna() -> retorna verdadeiro para linhas em que os valores não são NULL. É útil combonado a [] para realizar filtros

```python
Presenca = studant[studant["AttendanceRate"].notna()]

Presenca.head()
```

## Como seleciono linhas e colunas específicas de um dataFrame?


Eu estou interessando em saber a idade dos estudantes que estudam mais que 10 por semana.

loc -> operador para localizar informações por meio de filtro nos colchetes. Use-o para nome de colunas, rótulos de linhas ou expressão condicional

```python
estudar_mais_10 = studant.loc[studant["StudyHoursPerWeek"] > 15, "Age"]
#estudar_mais_10 = studant.loc[studant["StudyHoursPerWeek"] > 10, :] #Quando se deseja aplicar o filtro para todas as colunas

estudar_mais_10.head()
```

iloc -> usando para filtrar um determinado intervalor de linhas e colunas de seu interesse com base na posição.

```python
studant.iloc[8:15, 3:8] #Lembre-se que as posições começam em 0 tanto na linha quanto na coluna, logo, o filtro vai 8 a 14
```

Ao selecionar linhas e/ou colunas, é possível atribuir valores aos dados selecionados.

```python
studant.iloc[8:15, 6] = "Engineering"

studant.head(10)
```

## LEMBRE-SE

* Ao selecionar subconjuntos de dados, os colchetes [] são usados.
* Dentro desses colchetes, você pode usar um único rótulo de coluna/linha, uma lista de rótulos de colunas/linhas, um intervalo de rótulos, uma expressão condicional ou dois-pontos.
* Selecione linhas e/ou colunas específicas usando loc quando estiver utilizando os nomes das linhas e colunas.
* Selecione linhas e/ou colunas específicas usando iloc quando estiver utilizando as posições na tabela.
* Você pode atribuir novos valores a uma seleção com base em loc/iloc.
