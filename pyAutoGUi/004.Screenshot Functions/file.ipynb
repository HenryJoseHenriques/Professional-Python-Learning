## Others screenshot functions

PyAutoGUI oferece diversas funções para capturas de tela:

* screenshot(): Captura e retorna uma imagem do tipo Image (ou salva em arquivo) com a opção de capturar uma região específica da tela.
* Localização de Imagens: Funções como locateOnScreen() e locateCenterOnScreen() localizam elementos na tela comparando imagens. É possível ajustar a precisão com confidence.
* Correspondência de Pixels: Funções como pixelMatchesColor() verificam se um pixel específico corresponde a uma cor fornecida.
